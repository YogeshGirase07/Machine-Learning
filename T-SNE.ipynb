```python
# import libraries 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings("ignore")
```

```python
# load dataset
df = pd.read_csv("TSNE_data.csv")
```

```python
# Split-out validation dataset
array = df.values

# separate array into input and output components

X = array[:,1:]
Y = array[:,0]
```

```python
from sklearn.manifold import TSNE
from bioinfokit.visuz import cluster
```

```python
#TSNE Visualization

data_tsne =TSNE(n_components=2).fit_transform(X) # no. of components = 2
```

```python
data_tsne
```

```python
newdf = pd.DataFrame(data_tsne).join(df["diagnosis"])
newdf
```

```python
# get a list of categories
color_class = df['diagnosis'].to_numpy()
# convert dataframe to array. legnedpos = position on upper right corner showing color difference
cluster.tsneplot(score=data_tsne, colorlist=color_class, legendpos='upper right',legendanchor=(1.15,1))
plt.show()
#plot will be stored in the defalut directory
```
