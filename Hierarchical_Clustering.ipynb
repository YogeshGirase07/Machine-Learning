```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv('Universities.csv')
df
```

```python
# Customized / user defined Normalization function 
# here we can use build-in functions as well from sklearn but to show you how we can write customized function
def norm_func(i): # function name is norm_func, we can give any name here.
    x = (i-i.min())/(i.max()-i.min()) #(Xi-min)/range (Feature Scaling) ,Range =max-min
    return x                          # Apply Min-Max Normalization
                                    # Normalization lies between 0 to 1
```

```python
# (i-i.min())/(i.max()-i.min())
df['SAT'].min(),df['SAT'].max()
```

```python
(1310-1005)/(1415-1005)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(df.iloc[:,1:]) # from 1st column becouse 0 index col is univ names and we don't need
                                   #iloc means index location
```

```python
df_norm
```

```python
# Create dendrogram
dendrogram = sch.dendrogram(sch.linkage(df_norm,method="single"))
```

```python
# Create cluster , suppose got input from customer that go for 4 clusters
hc = AgglomerativeClustering(n_clusters=4, metric = 'euclidean', linkage= 'single')
```

```python
hc
```

```python
# save cluster for chart 
y_hc = hc.fit_predict(df_norm) # apply fit_predict method on dataset df_norm we will get cluster nos. in y_hc
cluster = pd.DataFrame(y_hc,columns=['Cluster_No']) # append those no. of cluster number create dataframe
```

```python
y_hc
```

```python
 cluster # Data point 0 belongs to 0th cluster, data point 1 belongs to 3rd cluster
```

```python
# Now let us the map this cluster membership to the data points
df['h_cluster'] = cluster
```
